```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
def read_tweets_with_before_after(isReply=False):
    path = './database/tweets_with_before_or_after.tsv'
    df = pd.read_csv(path, sep='\t')
    if isReply:
        df = df[df['tweet_text'].str.match('@(\w+) ')]
    else:
        df = df[df['tweet_text'].str.match('(?!@(\w+) )')]
    return df

df = read_tweets_with_before_after(isReply=False)
df.head(5)
```

```python
# xlm-roberta か Twitter BERT かを選択
# path = './database/emotion/estimate_result/tweet/xlm-roberta-base.tsv'
path = './database/emotion/estimate_result/tweet/Twitter-twhin-bert-base.tsv'
df_emo = pd.read_csv(path, sep='\t')
df_emo.head(5)
```

```python
# コロナ禍前後で感情の数を MBTI タイプ別にカウントする
def emotion_count_before_after_by_mbti(df):
    dic = {
        'Joy': 0,
        'Sadness': 1,
        'Anticipation': 2,
        'Surprise': 3,
        'Anger': 4,
        'Fear': 5,
        'Disgust': 6,
        'Trust': 7
    }
    before_or_after = ['before', 'after']
    
    for ba in before_or_after:
        emo_count_dic = {}
        df_ba = df[df['before_or_after'] == ba]
        tweet_ids = df_ba['tweet_id'].values
        for tw_id, emo in tqdm(zip(df_emo['tweet_id'], df_emo['emo'])):
            if tw_id in tweet_ids:
                if emo is not np.nan:
                    m_type = df[df['tweet_id'] == tw_id]['m_type_en'].values[0]
                    emotions = emo.split('|')
                    for e in emotions:
                        if m_type not in emo_count_dic:
                            emo_count_dic[m_type] = [0, 0, 0, 0, 0, 0, 0, 0]
                            emo_count_dic[m_type][dic[e]] = 1
                        else:
                            emo_count_dic[m_type][dic[e]] += 1
        print(ba)
        for i, j in sorted(emo_count_dic.items()):
            print(i, j, sep='\t')
emotion_count_before_after_by_mbti(df=df)
```

```python
# コロナ禍前後で MBTI タイプを考慮せずカウントする
def emotion_count_before_after(df):
    before_or_after = ['before', 'after']

    for ba in before_or_after:
        df_ba = df[df['before_or_after'] == ba]
        emo_count_dic = {}
        tweet_ids = df_ba['tweet_id'].values
        for tw_id, emo in tqdm(zip(df_emo['tweet_id'], df_emo['emo'])):
            if tw_id in tweet_ids:
                if emo is not np.nan:
                    emotions = emo.split('|')
                    for e in emotions:
                        if e not in emo_count_dic:
                            emo_count_dic[e] = 1
                        else:
                            emo_count_dic[e] += 1
        print(ba)
        for i, j in sorted(emo_count_dic.items()):
            print(i, j, sep='\t')
emotion_count_before_after(df=df)
```
